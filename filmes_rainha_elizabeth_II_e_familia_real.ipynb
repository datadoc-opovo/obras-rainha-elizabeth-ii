### dataframe 1

Pesquisa no IMDB: 'plot' contem 'Elizabet II'

https://www.imdb.com/search/title/?plot=%22Elizabeth%20II%22&count=250&view=advanced

Raspagem feita com Web Scraper. Sitemap: `sitemap_01_rainha_elizabeth_ii.txt`

```python
filmes_elizabeth = pd.read_csv('bases/rainha_elizabeth.csv')
```

```python
filmes_elizabeth = filmes_elizabeth[['titulo', 'titulo_ano', 'titulo_generos', 'titulo_avaliacao', 'titulo_votos',
                                     'titulo_resumo', 'data_lancamento', 'pais_origem', 'central_de_atendimento_oficial',
                                     'resumo_expandido', 'enredo', 'tipo', 'foto-src']]
```

```python
dict_tipo = {'Vídeo': 'Vídeo',
             'Série de TV': 'Série de TV',
             'Série Podcast': 'Série Podcast',
             'Curta para TV': 'Curta para TV',
             'Vídeo musical': 'Vídeo musical',
             'Especial de TV': 'Especial de TV',
             'Filme para televisão': 'Filme para televisão',
             'Minissérie de televisão': 'Minissérie de televisão'}

filmes_elizabeth['ds_tipo'] = filmes_elizabeth['tipo'].map(dict_tipo)
filmes_elizabeth['ds_tipo'] = filmes_elizabeth['ds_tipo'].fillna('não informado')
```

```python
filmes_elizabeth['titulo_generos_01'] = filmes_elizabeth['titulo_generos'].str.split(pat=',', expand=True)[0]
filmes_elizabeth['titulo_generos_02'] = filmes_elizabeth['titulo_generos'].str.split(pat=',', expand=True)[1]
filmes_elizabeth['titulo_generos_03'] = filmes_elizabeth['titulo_generos'].str.split(pat=',', expand=True)[2]

filmes_elizabeth_generos_1 = filmes_elizabeth[['titulo_generos_01', 'titulo']].value_counts().reset_index().rename(columns={
    'titulo_generos_01': 'genero',
    0: 'quantidade'
})

filmes_elizabeth_generos_2 = filmes_elizabeth[['titulo_generos_02', 'titulo']].value_counts().reset_index().rename(columns={
    'titulo_generos_02': 'genero',
    0: 'quantidade'
})

filmes_elizabeth_generos_3 = filmes_elizabeth[['titulo_generos_03', 'titulo']].value_counts().reset_index().rename(columns={
    'titulo_generos_03': 'genero',
    0: 'quantidade'
})
```

```python
filmes_elizabeth[['titulo_generos_01', 'titulo']].value_counts().reset_index()
```

```python
filmes_elizabeth_generos = pd.concat([filmes_elizabeth_generos_1, filmes_elizabeth_generos_2, filmes_elizabeth_generos_3])
filmes_elizabeth_generos['genero'] = filmes_elizabeth_generos['genero'].str.lstrip()
```

```python
# Visualização 1
filmes_elizabeth_generos.to_csv('arquivos_gerados/df_filmes_por_genero.csv', index=False)
```

```python
filmes_elizabeth_generos
```

```python
filmes_elizabeth['titulo_avaliacao'] = filmes_elizabeth['titulo_avaliacao'].astype(str).str.replace(',', '.').fillna(0.0).astype(float)
filmes_elizabeth['titulo_votos'] = filmes_elizabeth['titulo_votos'].fillna(0.0).astype(float)
```

```python
avaliacoes = filmes_elizabeth[['titulo', 'titulo_avaliacao', 'titulo_votos', 'titulo_resumo', 'foto-src']]
avaliacoes_validas = avaliacoes[(avaliacoes['titulo_avaliacao'] > 0.0) & (avaliacoes['titulo_votos'] > 0.0)]
```

```python
avaliacoes_validas
```

```python
coluna_avaliacao = []

for item in avaliacoes['titulo_avaliacao']:
    if item == 0.0:
        coluna_avaliacao.append('não foi avaliado')
    else:
        coluna_avaliacao.append('foi avaliado')
        
avaliacoes['avaliacao'] = coluna_avaliacao

# calcular percentual de avaliação dos títulos
situacao_avaliacoes = avaliacoes['avaliacao'].value_counts().reset_index().rename(columns={'index': 'situação'})
situacao_avaliacoes['percentual_avaliacao'] = round(situacao_avaliacoes['avaliacao']/situacao_avaliacoes['avaliacao'].sum()*100,2)
```

```python
situacao_avaliacoes
```

```python
# Visualização 2
df_avaliacoes_titulos = avaliacoes_validas.sort_values('titulo_avaliacao', ascending=False).reset_index()
df_avaliacoes_titulos = df_avaliacoes_titulos[['titulo', 'titulo_avaliacao', 'titulo_votos',  'foto-src']]
df_avaliacoes_titulos['foto-src'] = df_avaliacoes_titulos['foto-src'].fillna('sem foto')
```

```python
df_avaliacoes_titulos.to_csv('arquivos_gerados/df_avaliacoes_titulos.csv', index=False)
```

### dataset 2

Pesquisa: hashtag "reference-to-queen-elizabeth-ii"

https://www.imdb.com/search/keyword/?keywords=reference-to-queen-elizabeth-ii&ref_=kw_nxt&mode=detail&page=1&sort=moviemeter,asc

Raspagem feita com Web Scraper. Sitemap: `sitemap_02_rainha_elizabeth_ii.txt`

```python
referencias_rainha = pd.read_csv('bases/referencia_a_rainha_elizabeth_ii.csv')
```

```python
referencias_rainha = referencias_rainha[['numero', 'titulo', 'ano', 'estrelas', 'resumo', 'votos', 'nota', 'total', 'votos_avaliacao', 'foto-src', 'generos']]
```

```python
referencias_rainha['ano'] = referencias_rainha['ano'].str.split(pat='(', expand=True)[1]
referencias_rainha['ano'] = referencias_rainha['ano'].str.split(pat=')', expand=True)[0]
referencias_rainha['ano'] = referencias_rainha['ano'].str.split(pat='–', expand=True)[0]
referencias_rainha['ano'] = referencias_rainha['ano'].str.replace(' Music Video', '').str.replace(' TV Movie', '').str.replace(' TV Special', '').str.replace(' Video', '').str.replace('III', '').str.replace('I', '')
```

```python
referencias_rainha['generos_1'] = referencias_rainha['generos'].str.split(pat=',', expand=True)[0].str.lstrip()
referencias_rainha['generos_2'] = referencias_rainha['generos'].str.split(pat=',', expand=True)[1].str.lstrip()
referencias_rainha['generos_3'] = referencias_rainha['generos'].str.split(pat=',', expand=True)[2].str.lstrip()
```

```python
print('Segundo dados do IMDB, há ', referencias_rainha['titulo'].count(), 'títulos em que é feita referência à rainha Elizabeth II')
```

```python
df_generos_1 = referencias_rainha['generos_1'].value_counts().reset_index().rename(columns={'index': 'Gênero', 'generos_1': 'Quantidade'})
df_generos_2 = referencias_rainha['generos_2'].value_counts().reset_index().rename(columns={'index': 'Gênero', 'generos_2': 'Quantidade'})
df_generos_3 = referencias_rainha['generos_3'].value_counts().reset_index().rename(columns={'index': 'Gênero', 'generos_3': 'Quantidade'})
df_generos = pd.concat([df_generos_1, df_generos_2, df_generos_3])
```

```python
top10_generos_referencia = df_generos.groupby('Gênero').sum().sort_values('Quantidade', ascending=False).reset_index().head(10)
top10_generos_referencia
```

```python
top10_generos_referencia.to_csv('arquivos_gerados/top10_generos_referencia.csv', index=False)
```

```python
df_titulo_ano = referencias_rainha[referencias_rainha['ano'] != ''][['titulo', 'ano']].groupby('ano').count().sort_values('titulo', ascending=False).reset_index()
df_titulo_ano['percentual'] = round(df_titulo_ano['titulo']/df_titulo_ano['titulo'].sum()*100,2)
df_titulo_ano.head()
```

```python
print('Dos 203 títulos que fazem referência à rainha,', df_titulo_ano['titulo'].sum(), 'têm indicação do ano. A maioria é de 2021.')
```

```python
referencias_rainha['nota'] = referencias_rainha['nota'].str.replace(',', '.')
referencias_rainha['nota'] = referencias_rainha['nota'].fillna(0.0)
referencias_rainha['nota'] = referencias_rainha['nota'].astype(float)
```

```python
referencias_rainha_nota = referencias_rainha[['titulo', 'nota', 'votos_avaliacao', 'foto-src']].sort_values(by='nota', ascending=False).head(10)
referencias_rainha_nota.to_csv('arquivos_gerados/referencias_rainha_nota.csv', index=False)
```

### Concatenando as duas bases

```python
filmes_elizabeth_concat = pd.read_csv('bases/rainha_elizabeth.csv')
filmes_elizabeth_concat = filmes_elizabeth_concat[['titulo', 'titulo_ano', 'titulo_generos', 'foto-src']]
```

```python
filmes_elizabeth_concat['generos_1'] = filmes_elizabeth_concat['titulo_generos'].str.split(pat=',', expand=True)[0]
filmes_elizabeth_concat['generos_2'] = filmes_elizabeth_concat['titulo_generos'].str.split(pat=',', expand=True)[1]
filmes_elizabeth_concat['generos_3'] = filmes_elizabeth_concat['titulo_generos'].str.split(pat=',', expand=True)[2]
```

```python
filmes_elizabeth_concat['base'] = 'base 1'
```

```python
filmes_elizabeth_concat.rename(columns={'titulo_ano': 'ano', 'titulo_generos': 'generos'}, inplace=True)
```

```python
referencias_rainha_concat = pd.read_csv('bases/referencia_a_rainha_elizabeth_ii.csv')
referencias_rainha_concat = referencias_rainha_concat[['titulo', 'ano', 'generos', 'foto-src']]
```

```python
referencias_rainha_concat['generos_1'] = referencias_rainha_concat['generos'].str.split(pat=',', expand=True)[0]
referencias_rainha_concat['generos_2'] = referencias_rainha_concat['generos'].str.split(pat=',', expand=True)[1]
referencias_rainha_concat['generos_3'] = referencias_rainha_concat['generos'].str.split(pat=',', expand=True)[2]
```

```python
referencias_rainha_concat['base'] = 'base 2'
```

```python
df_bases_concatenadas = pd.concat([filmes_elizabeth_concat, referencias_rainha_concat]).reset_index()
df_bases_concatenadas.head()
```

```python
df_bases_concatenadas.drop_duplicates(subset='titulo', inplace=True)
```

```python
df_bases_concatenadas['novo_ano'] = df_bases_concatenadas['ano'].replace(regex=True,to_replace=r'\D',value=r'')
```

```python
df_bases_concatenadas['novo_ano_inicio'] = df_bases_concatenadas['novo_ano'].str.slice(start=0, stop=4)
```

```python
df_bases_concatenadas.head()
```

```python
df_titulos_groupby_ano = df_bases_concatenadas[['titulo', 'novo_ano_inicio']].groupby('novo_ano_inicio').count().sort_values('novo_ano_inicio', ascending=False).reset_index()
df_titulos_groupby_ano.to_csv('arquivos_gerados/df_titulos_groupby_ano.csv', index=False)
```

```python
df_titulos_groupby_ano
```

```python
df_bases_concatenadas[df_bases_concatenadas['titulo'] == 'The Crown']
```

```python
df_bases_concatenadas_genero1 = df_bases_concatenadas[['titulo', 'generos_1']].value_counts().reset_index().rename(columns={'generos_1': 'genero', 0: 'quantidade'})
df_bases_concatenadas_genero2 = df_bases_concatenadas[['titulo', 'generos_2']].value_counts().reset_index().rename(columns={'generos_2': 'genero', 0: 'quantidade'})
df_bases_concatenadas_genero3 = df_bases_concatenadas[['titulo', 'generos_3']].value_counts().reset_index().rename(columns={'generos_3': 'genero', 0: 'quantidade'})
```

```python
df_bases_concatenadas_total_genero = pd.concat([df_bases_concatenadas_genero1, df_bases_concatenadas_genero2, df_bases_concatenadas_genero3])
df_bases_concatenadas_total_genero['genero'] = df_bases_concatenadas_total_genero['genero'].str.lstrip()
```

```python
generos_obras = df_bases_concatenadas_total_genero[['genero', 'quantidade']].groupby('genero').count().sort_values('quantidade', ascending=False).reset_index()
df_top10_generos_obras = generos_obras.head(10)
```

```python
df_top10_generos_obras.to_csv('arquivos_gerados/df_top10_generos_obras.csv', index=False)
```
